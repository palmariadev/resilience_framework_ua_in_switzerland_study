# 1. Violin Plots by Gender

This notebook creates the final violin plots for the resilience variables grouped by `Gender`.

## Outputs
- Interactive Plotly displays in the notebook
- Standardized HTML exports in `outputs/html/`

```python
import logging
import sys
from pathlib import Path

from IPython.display import display

project_root = Path.cwd()
if project_root.name == "notebooks":
    project_root = project_root.parent
if str(project_root) not in sys.path:
    sys.path.append(str(project_root))

logging.basicConfig(level=logging.INFO, format="%(levelname)s: %(message)s")

from src.data_loader import load_survey_data
from src.plot_config import OUTPUT_HTML_DIR, RESILIENCE_LEVEL_COLUMNS
from src.violin_analysis import (
    build_violin_filename,
    create_gender_violin_figure,
    export_figure_html,
)

survey_df = load_survey_data()
OUTPUT_HTML_DIR.mkdir(parents=True, exist_ok=True)
survey_df.head()
```

```python
# Plot configuration block
plot_prefix = "violin_gender"
output_directory = OUTPUT_HTML_DIR

for level_column in RESILIENCE_LEVEL_COLUMNS:
    figure, summary_df, anova_results = create_gender_violin_figure(survey_df, level_column)
    export_path = export_figure_html(
        figure,
        build_violin_filename(plot_prefix, level_column),
    )

    display(summary_df)
    display({
        "variable": level_column,
        "f_statistic": round(anova_results["f_statistic"], 3),
        "p_value": round(anova_results["p_value"], 3),
        "export_path": str(export_path),
    })
    figure.show()
```
