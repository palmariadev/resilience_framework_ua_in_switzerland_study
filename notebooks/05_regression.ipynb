# 5. Regression

This notebook runs the curated resilience regression workflow and exports the final regression visuals.

## Outputs
- Interactive Plotly figures in the notebook
- Standardized HTML exports in `outputs/html/`

```python
import logging
import sys
from pathlib import Path

from IPython.display import display

project_root = Path.cwd()
if project_root.name == "notebooks":
    project_root = project_root.parent
if str(project_root) not in sys.path:
    sys.path.append(str(project_root))

logging.basicConfig(level=logging.INFO, format="%(levelname)s: %(message)s")

from src.data_loader import load_survey_data
from src.regression_analysis import (
    create_feature_importance_heatmap,
    create_model_performance_figure,
    export_regression_figures,
    fit_resilience_models,
)

survey_df = load_survey_data()
survey_df.head()
```

```python
# Plot configuration block
performance_title = "Regression Performance Across Resilience Outcomes"
heatmap_title = "Regression Feature Importance by Outcome"
figure_width = 1000
figure_height = 700

model_results, summary_df = fit_resilience_models(survey_df)
display(summary_df)

if summary_df.empty:
    raise ValueError("No regression models were fitted. Check predictor availability.")

performance_figure = create_model_performance_figure(summary_df)
importance_figure = create_feature_importance_heatmap(model_results)
export_regression_figures(performance_figure, importance_figure)

performance_figure.show()
importance_figure.show()
```
